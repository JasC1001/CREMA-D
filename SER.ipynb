```python
import os
import numpy as np
import pandas as pd
import tensorflow as tf
import librosa
import librosa.display
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import classification_report, confusion_matrix
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv1D, MaxPooling1D, Flatten, Dense, BatchNormalization, ReLU, Softmax, Input
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.optimizers.schedules import LearningRateSchedule
from tqdm import tqdm

# --- 1. DATA LOADING AND PREPROCESSING ---

# Define RAVDESS and CREMA-D mappings
ravdess_map = { "01": "neutral",
                "02": "calm",
                "03": "happy",
                "04": "sad",
                "05": "angry",
                "06": "fearful",
                "07": "disgust",
                "08": "surprise" }

cremad_map = { "NEU": "neutral",
               "HAP": "happy",
               "SAD": "sad",
               "ANG": "angry",
               "FEA": "fearful",
               "DIS": "disgust" }

def add_noise(audio, noise_level=0.001):
    noise = np.random.normal(0, noise_level, len(audio))
    return audio + noise

def shift_pitch(audio, sr, n_steps=-0.25):
    return librosa.effects.pitch_shift(y=audio, sr=sr, n_steps=n_steps)

def load_and_augment_cremad(data_dir):
    X, y = [], []
    fixed_length = 216  # Fixed time steps for MFCC
    files = [os.path.join(root, file) for root, _, filenames in os.walk(data_dir) for file in filenames if file.endswith(".wav")]
    for file_path in tqdm(files, desc="Loading CREMA-D Dataset"):
        try:
            parts = os.path.basename(file_path).split("_")
            emotion_code = parts[2]
            emotion = cremad_map.get(emotion_code)
            if emotion:
                audio, sr = librosa.load(file_path, sr=22050, duration=2.5)
                mfcc = librosa.feature.mfcc(y=audio, sr=sr, n_mfcc=20)
                mfcc = np.pad(mfcc, ((0, 0), (0, max(0, fixed_length - mfcc.shape[1]))), mode='constant')[:, :fixed_length]
                X.append(mfcc.T)
                y.append(emotion)

                # Data augmentation
                audio_noise = add_noise(audio)
                mfcc_noise = librosa.feature.mfcc(y=audio_noise, sr=sr, n_mfcc=20, n_fft=2048, hop_length=512)
                mfcc_noise = np.pad(mfcc_noise, ((0, 0), (0, max(0, fixed_length - mfcc_noise.shape[1]))), mode='constant')[:, :fixed_length]
                X.append(mfcc_noise.T)
                y.append(emotion)

                audio_pitch = shift_pitch(audio, sr)
                mfcc_pitch = librosa.feature.mfcc(y=audio_pitch, sr=sr, n_mfcc=20, n_fft=2048, hop_length=512)
                mfcc_pitch = np.pad(mfcc_pitch, ((0, 0), (0, max(0, fixed_length - mfcc_pitch.shape[1]))), mode='constant')[:, :fixed_length]
                X.append(mfcc_pitch.T)
                y.append(emotion)

        except Exception as e:
            print(f"Error processing {file_path}: {e}")
    return np.array(X), np.array(y)

def load_and_augment_ravdess(data_dir):
    X, y = [], []
    fixed_length = 216
    files = [os.path.join(root, file) for root, _, filenames in os.walk(data_dir) for file in filenames if file.endswith(".wav")]
    for file_path in tqdm(files, desc="Loading RAVDESS Dataset"):
        try:
            parts = os.path.basename(file_path).split("-")
            emotion_code = parts[2]
            emotion = ravdess_map.get(emotion_code)
            if emotion:
                audio, sr = librosa.load(file_path, sr=22050, duration=2.5)
                mfcc = librosa.feature.mfcc(y=audio, sr=sr, n_mfcc=20)
                mfcc = np.pad(mfcc, ((0, 0), (0, max(0, fixed_length - mfcc.shape[1]))), mode='constant')[:, :fixed_length]
                X.append(mfcc.T)
                y.append(emotion)

                # Data augmentation
                audio_noise = add_noise(audio)
                mfcc_noise = librosa.feature.mfcc(y=audio_noise, sr=sr, n_mfcc=20, n_fft=2048, hop_length=512)
                mfcc_noise = np.pad(mfcc_noise, ((0, 0), (0, max(0, fixed_length - mfcc_noise.shape[1]))), mode='constant')[:, :fixed_length]
                X.append(mfcc_noise.T)
                y.append(emotion)

                audio_pitch = shift_pitch(audio, sr)
                mfcc_pitch = librosa.feature.mfcc(y=audio_pitch, sr=sr, n_mfcc=20, n_fft=2048, hop_length=512)
                mfcc_pitch = np.pad(mfcc_pitch, ((0, 0), (0, max(0, fixed_length - mfcc_pitch.shape[1]))), mode='constant')[:, :fixed_length]
                X.append(mfcc_pitch.T)
                y.append(emotion)
        except Exception as e:
            print(f"Error processing {file_path}: {e}")
    return np.array(X), np.array(y)

# --- 2. MODEL TRAINING AND EVALUATION ---
def build_model(input_shape, num_classes):
    model = Sequential([
        Input(shape=input_shape),
        Conv1D(512, kernel_size=5, strides=1, padding='same', activation='relu'),
        BatchNormalization(),
        MaxPooling1D(pool_size=5, strides=2, padding='same'),

        Conv1D(512, kernel_size=5, strides=1, padding='same', activation='relu'),
        BatchNormalization(),
        MaxPooling1D(pool_size=5, strides=2, padding='same'),

        Conv1D(256, kernel_size=5, strides=1, padding='same', activation='relu'),
        BatchNormalization(),
        MaxPooling1D(pool_size=5, strides=2, padding='same'),

        Conv1D(256, kernel_size=3, strides=1, padding='same', activation='relu'),
        BatchNormalization(),
        MaxPooling1D(pool_size=5, strides=2, padding='same'),

        Conv1D(128, kernel_size=3, strides=1, padding='same', activation='relu'),
        BatchNormalization(),
        MaxPooling1D(pool_size=5, strides=2, padding='same'),

        Flatten(),
        Dense(512, activation='relu'),
        BatchNormalization(),
        Dense(num_classes, activation='softmax')
    ])
    return model

def plot_emotion_counts(y_cremad, y_ravdess):
    # Count emotion labels for each dataset
    cremad_counts = pd.Series(y_cremad).value_counts()
    ravdess_counts = pd.Series(y_ravdess).value_counts()

    # Create subplots for CREMA-D and RAVDESS
    plt.figure(figsize=(12, 5))

    # CREMA-D emotion counts
    plt.subplot(1, 2, 1)
    sns.barplot(x=cremad_counts.index, y=cremad_counts.values, palette='Set2')
    plt.title("Emotions Counts of CREMA-D")
    plt.xlabel("Emotion")
    plt.ylabel("Count")
    for i, count in enumerate(cremad_counts.values):
        plt.text(i, count + 50, str(count), ha='center')

    # RAVDESS emotion counts
    plt.subplot(1, 2, 2)
    sns.barplot(x=ravdess_counts.index, y=ravdess_counts.values, palette='Set3')
    plt.title("Emotions Counts of RAVDESS")
    plt.xlabel("Emotion")
    plt.ylabel("Count")
    for i, count in enumerate(ravdess_counts.values):
        plt.text(i, count + 10, str(count), ha='center')

    plt.tight_layout()
    plt.show()
    plt.savefig("emotion_counts.png")
    print("Emotion counts saved to emotion_counts.png")

def train_and_evaluate(X, y, dataset_name):
    label_encoder = LabelEncoder()
    y_encoded = label_encoder.fit_transform(y)
    X_train, X_test, y_train, y_test = train_test_split(X, y_encoded, test_size=0.2, random_state=42)

    input_shape = (X_train.shape[1], X_train.shape[2])
    num_classes = len(np.unique(y_train))
    model = build_model(input_shape, num_classes)

    optimizer = tf.keras.optimizers.Adam(
        learning_rate=0.001,
        weight_decay=5e-4
    )

    model.compile(optimizer=optimizer, loss='sparse_categorical_crossentropy', metrics=['accuracy'])
    checkpoint = tf.keras.callbacks.ModelCheckpoint(f"best_model_{dataset_name}.keras", save_best_only=True, monitor="val_accuracy", mode="max")

    history = model.fit(X_train, y_train, validation_data=(X_test, y_test), epochs=50, batch_size=64, callbacks=[checkpoint])

    # 保存最後訓練的模型權重
    model.save_weights(f"final_model_weights_{dataset_name}.weights.h5")
    print(f"Final model weights saved to final_model_weights_{dataset_name}.weights.h5")

    # Evaluate
    loss, accuracy = model.evaluate(X_test, y_test)
    print(f"{dataset_name} Accuracy: {accuracy * 100:.2f}%")

    # 保存完整模型
    model.save(f"final_model_{dataset_name}.keras", save_format="keras")
    print(f"Full model saved to final_model_{dataset_name}.keras")

    # Classification report
    y_pred = np.argmax(model.predict(X_test), axis=1)
    print(classification_report(y_test, y_pred, target_names=label_encoder.classes_))

    # Confusion Matrix
    cm = confusion_matrix(y_test, y_pred)
    plt.figure(figsize=(8, 6))
    sns.heatmap(cm, annot=True, fmt='d', cmap='Reds', xticklabels=label_encoder.classes_, yticklabels=label_encoder.classes_)
    plt.title(f'Confusion Matrix - {dataset_name}')
    plt.xlabel('Predicted')
    plt.ylabel('Actual')
    plt.show()
    plt.savefig(f"confusion_matrix_{dataset_name}.png")
    print(f"Confusion Matrix saved to confusion_matrix_{dataset_name}.png")

    # Plot accuracy and loss curves
    plt.figure(figsize=(12, 4))
    plt.subplot(1, 2, 1)
    plt.plot(history.history['loss'], label='Train Loss')
    plt.plot(history.history['val_loss'], label='Validation Loss')
    plt.title(f'{dataset_name} Loss Curve')
    plt.xlabel('Epochs')
    plt.ylabel('Loss')
    plt.legend()

    plt.subplot(1, 2, 2)
    plt.plot(history.history['accuracy'], label='Train Accuracy')
    plt.plot(history.history['val_accuracy'], label='Validation Accuracy')
    plt.title(f'{dataset_name} Accuracy Curve')
    plt.xlabel('Epochs')
    plt.ylabel('Accuracy')
    plt.legend()
    plt.show()
    plt.savefig(f"accuracy_loss_curve_{dataset_name}.png")
    print(f"Accuracy and Loss Curves saved to accuracy_loss_curve_{dataset_name}.png")

# --- 3. MAIN EXECUTION ---
RAVDESS_PATH = "Audio_Speech_Actors"
CREMAD_PATH = "CREMA-D/AudioWAV"

print("Loading and Augmenting RAVDESS Dataset...")
X_ravdess, y_ravdess = load_and_augment_ravdess(RAVDESS_PATH)
print(f"Loaded RAVDESS data: {X_ravdess.shape[0]} samples")

print("Loading and Augmenting CREMA-D Dataset...")
X_cremad, y_cremad = load_and_augment_cremad(CREMAD_PATH)
print(f"Loaded CREMA-D data: {X_cremad.shape[0]} samples")

print("Plotting emotion distributions...")
plot_emotion_counts(y_cremad, y_ravdess)

train_and_evaluate(X_ravdess, y_ravdess, "RAVDESS")
train_and_evaluate(X_cremad, y_cremad, "CREMA-D")
```

```python
import os
import numpy as np
import librosa
from tensorflow.keras.models import load_model
from sklearn.preprocessing import LabelEncoder
import matplotlib.pyplot as plt
import librosa.display

# --- FUNCTION DEFINITIONS ---

def predict_emotion(model, file_path, label_encoder, fixed_length=216):
    """Predicts the emotion of a given audio file and saves its MFCC visualization."""
    try:
        # Load the audio file
        audio, sr = librosa.load(file_path, sr=22050, duration=2.5)

        # Compute MFCCs
        mfcc = librosa.feature.mfcc(y=audio, sr=sr, n_mfcc=20)
        mfcc = np.pad(mfcc, ((0, 0), (0, max(0, fixed_length - mfcc.shape[1]))), mode='constant')[:, :fixed_length]
        mfcc = np.expand_dims(mfcc.T, axis=0)  # Expand dimensions to match model input

        # Save MFCC visualization
        plt.figure(figsize=(10, 4))
        librosa.display.specshow(mfcc[0].T, x_axis='time', sr=sr, cmap='viridis')
        plt.colorbar(format='%+2.0f dB')
        plt.title('MFCC')
        plt.tight_layout()
        output_image_path = file_path.replace('.wav', '_mfcc.png')
        plt.savefig(output_image_path)
        plt.close()
        print(f"MFCC visualization saved to {output_image_path}")

        # Predict
        predictions = model.predict(mfcc)
        predicted_label_index = np.argmax(predictions, axis=1)[0]
        predicted_emotion = label_encoder.inverse_transform([predicted_label_index])[0]

        # Print and return result
        print(f"Predicted Emotion: {predicted_emotion}")
        return predicted_emotion

    except Exception as e:
        print(f"Error processing the file: {e}")
        return None

# --- MAIN EXECUTION ---
if __name__ == "__main__":
    # Select dataset model
    dataset_choice = input("Select dataset model (RAVDESS or CREMA-D): ").strip().upper()
    if dataset_choice not in ["RAVDESS", "CREMA-D"]:
        print("Invalid dataset choice. Please select 'RAVDESS' or 'CREMA-D'.")
        exit(1)

    # Load the trained model based on dataset choice
    model_path = f"final_model_{dataset_choice}.keras"  # Replace with your trained model paths
    if not os.path.exists(model_path):
        print(f"Model file not found at {model_path}.")
        exit(1)

    model = load_model(model_path)

    # Load the label encoder for the chosen dataset
    label_encoder = LabelEncoder()
    if dataset_choice == "RAVDESS":
        label_encoder.fit(["neutral", "calm", "happy", "sad", "angry", "fearful", "disgust", "surprise"])
    elif dataset_choice == "CREMA-D":
        label_encoder.fit(["neutral", "happy", "sad", "angry", "fearful", "disgust"])

    # Input audio file for testing
    test_file_path = input("Enter the path to the .wav file: ").strip()
    if not os.path.exists(test_file_path):
        print(f"Audio file not found at {test_file_path}.")
        exit(1)

    # Predict the emotion
    predicted_emotion = predict_emotion(model, test_file_path, label_encoder)
```
